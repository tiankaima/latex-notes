```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
mass = [0,100.05, 100.06, 99.81, 99.99, 99.94]
total_mass = []
for _m in mass:
    total_mass.append(_m + total_mass[-1] if len(total_mass) else _m)

print(total_mass)
```

```python
length = [2.199,2.321,2.518,2.703,2.908,3.323]
voltage = [0.001, -0.034, -0.042,-0.196, -0.522, -0.657]
```

```python
plt.plot(length, voltage)
(k,b) = np.polyfit(length, voltage, 1)
print(k,b)
plt.plot(length, k*np.array(length) + b)
plt.savefig("./img/Fig.2.png")
```

```python
plt.plot(total_mass, length)
(k,b) = np.polyfit(total_mass, length, 1)
print(k,b)
plt.plot(total_mass, k*np.array(total_mass) + b)
plt.savefig("./img/Fig.3.png")
```
