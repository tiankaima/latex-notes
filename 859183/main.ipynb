```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s1 = np.array([2.522, 2.526, 2.524, 2.524, 2.521, 2.522,
              2.517, 2.514, 2.512, 2.509, 2.444, 2.310])
o1 = np.array([2.517, 2.514, 2.510, 2.458, 2.399, 2.289,
              2.100, 1.937, 1.798, 1.679, 0.422, 0.229])
rs = np.array([100, 500, 1000, 5000, 10000, 20000, 40000,
              60000, 80000, 100000, 1000000, 2000000]) / 2
r_g = 100000
```

```python
l1, = plt.plot(r_g/rs, (s1 - o1)/s1, 'o', label='data')
l2, = plt.plot(r_g/rs, 1/(1+r_g/rs), 'r', label='prediction')
plt.legend(handles=[l1, l2, ], labels=['data', 'theory'], loc=0)
plt.savefig('img/Fig.4.png')
```
